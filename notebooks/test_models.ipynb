```python
from src.models.mlp_classifier import MLPClassifier
from src.preprocessing.io.feature_loader import FeatureLoader
from src.common.constants import Constants as consts
feature_loader = FeatureLoader(file_name=consts.feature_extracted, feat_suffix="fft")

train_meta, train_embeddings = feature_loader.load_data_split(split_name="train")
print(train_meta.shape, train_embeddings.shape)
```

```python
sampled_train_meta, sampled_train_emb = feature_loader.sample_fraction_metadata_and_embeddings(train_meta, train_embeddings, fraction=0.4)
print(sampled_train_meta.shape, sampled_train_emb.shape)
```

```python
dev_meta, dev_embeddings = feature_loader.load_data_split(split_name="dev")
print(dev_meta.shape, dev_embeddings.shape)
sampled_dev_meta, sampled_dev_emb = feature_loader.sample_fraction_metadata_and_embeddings(dev_meta, dev_embeddings, fraction=0.2)
print(sampled_dev_meta.shape, sampled_dev_emb.shape)
```

```python
# Test on FFT embeddings without cluster ids 10%
clf = MLPClassifier()
clf.optuna_fit(
    n_trials=50,
    X_train=sampled_train_emb,
    y_train=sampled_train_meta['target'].to_numpy(),
    X_dev=sampled_dev_emb,
    y_dev=sampled_dev_meta['target'].to_numpy(),
)
print("Best trial value (lowest loss):", clf.get_best_value())
```

```python
# Test on FFT embeddings without cluster ids 10%
clf = MLPClassifier()
clf.optuna_fit(
    n_trials=50,
    X_train=sampled_train_emb,
    y_train=sampled_train_meta['target'].to_numpy(),
    X_dev=sampled_dev_emb,
    y_dev=sampled_dev_meta['target'].to_numpy(),
)
print("Best trial value (lowest loss):", clf.get_best_value())
```

```python
# Test on FFT embeddings without cluster ids
clf = MLPClassifier()
clf.optuna_fit(
    n_trials=50,
    X_train=sampled_train_emb,
    y_train=sampled_train_meta['target'].to_numpy(),
    X_dev=sampled_dev_emb,
    y_dev=sampled_dev_meta['target'].to_numpy(),
    train_cluster_id=sampled_train_meta['cluster_id'].to_numpy(),
    dev_cluster_id=sampled_dev_meta['cluster_id'].to_numpy()
)
print("Best trial value (lowest loss):", clf.get_best_value())
```
