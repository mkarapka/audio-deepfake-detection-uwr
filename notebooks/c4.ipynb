```python
import time
def iterate_through_ds(iters, ds):
  start = time.time()
  for i, _ in enumerate(ds):
    if i == iters:
      break

  return time.time() - start
```

```python
from datasets import load_dataset

mls_ds = load_dataset("parler-tts/mls_eng", split="test")
audeter_ds = load_dataset("wqz995/AUDETER", "mls-tts-bark", split='dev', streaming=True)
```

```python
RECORDS = 1000
print("Time for mls 1000 records:", iterate_through_ds(iters=RECORDS, ds=mls_ds))
```

```python
RECORDS = 1000
print("Time for audeter 1000 records:", iterate_through_ds(iters=RECORDS, ds=audeter_ds))
```
