```python
from src.preprocessing.io.feature_loader import FeatureLoader
from src.common.constants import Constants as consts

feature_loader = FeatureLoader(feat_suffix=consts.wavlm_emb_suffix)
meta_train, feat_train = feature_loader.load_data_split(split_name="train")
print(meta_train.shape, feat_train.shape)
```

```python
print(meta_train['config'].value_counts())
print(meta_train['config'].value_counts().count())
```

```python
records = meta_train.query('config.str.contains("vocoders")')
summary = records["config"].value_counts()

print(summary.sum())
print(
    meta_train["config"].value_counts().sum()
    - meta_train.query('config.str.contains("tts")').value_counts().sum()
    - summary.sum()
)
```

```python
vocs = 'config.str.contains("vocoders")'
tts = 'config.str.contains("tts")'
```

```python
vocoders_tts = meta_train.query(f"{vocs} or {tts}")
print(vocoders_tts['config'].value_counts())
```
