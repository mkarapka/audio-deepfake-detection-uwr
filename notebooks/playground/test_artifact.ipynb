```python
import wandb
from src.common.wandb_config import WANDB_ENTITY, WANDB_PROJECT
import joblib
```

```python
params_name = "LogisticRegressionObjective_fft_test_best_params"
```

```python
from pathlib import Path

run_test = wandb.init(
    project=WANDB_PROJECT,
    entity=WANDB_ENTITY,
    name="test-run",
    tags=["test", "playground"],
)
artifact = run_test.use_artifact(f"{params_name}:latest", type="model_params")

artifact_dir = Path(artifact.download())
print("Artifact downloaded to:", artifact_dir)

params = joblib.load(artifact_dir / f"{params_name}.pkl")
print("Loaded parameters:", params)
```

```python
run_test.finish()
```
